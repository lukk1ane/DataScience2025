```python
import numpy as np
```

```python
# Task 1

arr1 = np.arange(1, 10)
print(arr1)
print(np.sum(arr1))
print(np.prod(arr1))
```

```python
# Task 2

arr2 = np.arange(0, 20)
print(arr2[3:8])
```

```python
# Task 3
arr2 = np.arange(0, 25)
arr2 = arr2.reshape(5,5)
print(arr2)
print(arr2[0:2, :])
print(arr2[:, -2:])
```

```python
# Task 4
a = np.array([1,2,3,4])
b = np.array([2,4,6,8])
print(a + b)
print(a * b)
print(np.sin(b))
```

```python
# Task 5
from random import randint

arr = np.array([randint(1, 50) for i in range(0, 20)])
arr = np.random.randint(1, 51, size=20)
mask = (arr > 25) & (arr < 40)

arr2 = arr[mask]
print(arr2)
```

```python
# Task 6

array = np.arange(25).reshape(5, 5)
np.fill_diagonal(array, 0)
array = np.where(array != 0, array * 2, 0)
print(array)
```

```python
# Task 7
array = np.arange(9).reshape(3,3)
print(array[2, :])
print(array[:, :2])
```

```python
# Task 8
array = np.arange(1, 7).reshape(2,3)
print(array, "\n",np.transpose(array))
```

```python
# Task 9
array1 = np.arange(1, 10).reshape(3,3)
array2 = np.arange(1, 4).reshape(1,3)
print(array1, array2, array2+array1)
```
