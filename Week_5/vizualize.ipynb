### Task3

```python
df = sns.load_dataset("iris")


df.head()

species_unique = df['species'].unique()

setosa      = df[df["species"] == "setosa"]
versicolor  = df[df["species"] == "versicolor"]
virginica   = df[df["species"] == "virginica"]

plt.figure(figsize=(20,8))

plt.scatter(setosa['petal_length'], setosa['petal_width'], marker='o', label = "setosa")
plt.scatter(versicolor['petal_length'], versicolor['petal_width'], marker='x', label='versicolor')
plt.scatter(virginica['petal_length'],virginica['petal_width'], marker="^", label="virginica")

plt.xlabel('length')
plt.ylabel('width')
plt.title('length VS width')
plt.legend()
plt.show()


```

### Task4

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
store_A = [1200, 1400, 1500, 1600, 1700, 1800]
store_B = [1100, 1300, 1250, 1400, 1600, 1650]
store_C = [900, 1200, 1100, 1350, 1500, 1600]

x = np.arange(len(months))
width = 0.25  # width of each bar

plt.figure(figsize=(10,5))



plt.bar(x - width, store_A, width, label='Store A')
plt.bar(x,         store_B, width, label='Store B')
plt.bar(x + width, store_C, width, label='Store C')

plt.xlabel("Month")
plt.ylabel("Average Monthly Revenue ($)")
plt.title("Monthly Revenue Comparison Across 3 Stores")
plt.xticks(x, months)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### task 8

```python
df = pd.DataFrame(np.reshape(np.random.randint(100, 1000, 120), (60, 2)),
                  columns=['Product1', 'Product2'])

# x = days/months index
x = df.index
sales = df['Product1']  # only Product1

plt.figure(figsize=(10,4))
plt.plot(x, sales, marker='o', linewidth=2, label='Product 1')

plt.xlabel('Day')
plt.ylabel('Sales')
plt.title('Daily Sales of Product 1')
plt.grid(True)
plt.legend()

# highlight maximum & annotate
mx_idx = sales.idxmax()
plt.scatter(x[mx_idx], sales[mx_idx], s=120, edgecolor='red', facecolor='none')
plt.annotate(f'Max: {sales[mx_idx]}',
             xy=(x[mx_idx], sales[mx_idx]),
             xytext=(x[mx_idx] + 1, sales[mx_idx] + 30),
             arrowprops=dict(arrowstyle='->'))

plt.tight_layout()
plt.show()
```

### task 10

```python
dates = pd.date_range('2024-01-01', periods=30)
p1 = np.random.randint(20,120, size=30)
p2 = np.random.randint(10,100, size=30)
revenue = p1*5 + p2*3

fig, ax = plt.subplots(2,2, figsize=(12,8))
# Line
ax[0,0].plot(dates, p1, label='Product1')
ax[0,0].plot(dates, p2, label='Product2')
ax[0,0].set_title('Daily Units Sold')
ax[0,0].legend()
ax[0,0].tick_params(axis='x', rotation=45)

# Bar
ax[0,1].bar(['P1','P2'], [p1.sum(), p2.sum()])
ax[0,1].set_title('Total Units Sold (month)')

# Histogram of revenue
ax[1,0].hist(revenue, bins=15)
ax[1,0].set_title('Revenue Distribution')

# Pie of contribution to revenue
labels = ['Product1','Product2']
tot = [p1.sum()*5, p2.sum()*3]
ax[1,1].pie(tot, labels=labels, autopct='%1.1f%%', startangle=90)
ax[1,1].set_title('Revenue Share')

plt.tight_layout()
plt.show()
```
