```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = np.arange(1, 8)


plt.plot(days, temperatures, marker='o', color='orange', linestyle='--')
plt.xlabel("Day of the Week")
plt.ylabel("Temperature (Â°C)")
plt.title("Temperature Changes Over a Week")
plt.grid(True)
```

```python
days = np.arange(1, 31)
product_a = np.random.randint(20, 50, size=30)
product_b = np.random.randint(15, 60, size=30)

plt.plot(days, product_a, label="Product 1", marker="o")
plt.plot(days, product_b, label="Product 2", marker="s")
plt.xlabel("Month")
plt.ylabel("Sold")
plt.title("comparison")
plt.legend()
plt.grid(True)
```

```python
months = ["Jan", "Feb", "Mar", "Apr"]
store_1 = [2000, 2500, 2200, 2700]
store_2 = [1800, 2300, 2100, 2600]
store_3 = [2200, 2400, 2000, 2800]

x = np.arange(len(months))
width = 0.25

plt.bar(x - width, store_1, width, label="Store 1")
plt.bar(x, store_2, width, label="Store 2")
plt.bar(x + width, store_3, width, label="Store 3")

plt.xlabel("Month")
plt.ylabel("Revenue ($)")
plt.title("Average Monthly Revenue per Store")
plt.xticks(x, months)
plt.legend();
```

```python
data1 = np.random.normal(50, 10, 1000)
data2 = np.random.normal(60, 15, 1000)

plt.hist(data1, bins=20, alpha=0.6, label="Dataset 1", color="blue")
plt.hist(data2, bins=20, alpha=0.8, label="Dataset 2", color="red")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Comparison")
plt.legend()
```

```python
segments = [40, 25, 20, 15]
labels = ["Segment A", "Segment B", "Segment C", "Segment D"]
explode = (0.2, 0, 0, 0)  # Highlight first segment

plt.pie(segments, labels=labels, explode=explode, autopct="%1.1f%%", shadow=True)
plt.title("Customer Segment Percentages");
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperatures = np.arange(1, 8)  # 7 values

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
print(fig)
# Line plot
axes[0, 0].plot(days, temperatures, marker="o", color="orange")
axes[0, 0].set_title("Line Plot")

# Bar chart
axes[0, 1].bar(["A", "B", "C"], [10, 20, 15], color="green")
axes[0, 1].set_title("Bar Chart")

# Histogram
axes[1, 0].hist(np.random.randn(100), bins=10, color="purple")
axes[1, 0].set_title("Histogram")

# Scatter plot
axes[1, 1].scatter(np.random.rand(20), np.random.rand(20), color="red")
axes[1, 1].set_title("Scatter Plot")

plt.tight_layout()
```

```python
cust = [22, 24, 21, 23, 25, 26, 24]

plt.plot(days, cust, marker="o", color="blue", linestyle="-")
plt.title("Weekly Temperature with Annotations")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")


max_temp = max(cust)
max_day = days[cust.index(max_temp)]
plt.annotate(
    f"Max: {max_temp}Â°C",
    xy=(max_day, max_temp),
    xytext=(max_day, max_temp + 1),
    arrowprops=dict(facecolor="red", shrink=0.05),
)

plt.grid(True)
```
