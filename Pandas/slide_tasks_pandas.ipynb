```python
data = pd.Series([5, 2, 3, 0], index=['a', 'b', 'c', 'd'])

data1 = pd.Series([1,6,4,9], index = ['a', 'b','d','e'])
print(data, "\n\n", data1)


result_add = data.add(data1, fill_value=0)
print("Addition \n", result_add)

result_sub = data.sub(data1, fill_value=0)
print("Subtraction \n", result_sub)

result_mul = data.mul(data1)
result_mul = result_mul.dropna()
print("Multiplication \n", result_mul)

result_div = data.div(data1)
result_div = result_div.dropna()
print("Division \n", result_div)

```

```python
data = pd.read_csv("orders.csv")
data.dropna(inplace=True)

data_before = data.dtypes

data["Product"] = data["Product"].astype("string")
data["Price"] = data["Price"].astype("int")

after = data.dtypes
print("Before conversion \n", data_before)
print("After conversion \n", after)
```
