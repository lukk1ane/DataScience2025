```python
import numpy as np
a = np.arange(25).reshape(5, 5)

print(a[:, 1::2])
print(a[1::2, :3:2])
print(a[-1, ::])


```

```python
a = np.array([10, 20, 30, 40, 50])
mask = np.array([0, 1, 0, 1, 0], dtype = bool)
y = a[mask]
print(y)
mask2 = a > 25
y2 = a[mask2]
print(y2)
```

```python
a = np.arange(25).reshape(5, 5)
mask = a % 3 == 0
print(a)
print(a[[0, 1, 2, 3], [1, 2, 3, 4]])
print(np.argmax(a))
print(np.unravel_index(np.argmax(a), a.shape))
```

```python
arr = np.arange(-15, 15).reshape(5, 6) ** 2
print(arr.max(axis=1))
print(arr.mean(axis=0, dtype = np.int32))
print(np.unravel_index(arr.argmin(), arr.shape), arr.argmin())

```
