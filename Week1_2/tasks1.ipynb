```python
# Task 6
fruits = ["apple", "banana", "orange"]
fruit_colors = {"apple":"red", "banana":"yellow", "orange":"orange"}

for fruit in fruits:
    print(f"The fruit {fruit} is {fruit_colors[fruit]}")

for fruit_key in fruit_colors:
    print(fruit_key)
for fruit_values in fruit_colors.values():
    print(fruit_values)
for fruit_both in fruit_colors.items():
    print(fruit_both[::-1])
```

```python
# Task 9

def fizz_buzz(n):
    for i in range (1, n):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 5 == 0:
            print("Buzz")
        elif i % 3 == 0:
            print("Fizz")
        else: 
            print(i)
fizz_buzz(20)
```

```python
# Task 10
def division(x, y):
    try: 
        return x / y
    except ZeroDivisionError:
        print("Can not divide by zero sir")
print(division(2,3))
print(division(3,0))
```

```python
# Task 11
def look_up(list, index):
    try:
        return list[index]
    except IndexError:
        print("Index out of bounds!")
        return
    except TypeError:
        print("Incorrect types provided")
        return
print(look_up([1, 2, 3], 0))
print(look_up([1, 2, 3], 5))
print(look_up(5, 0))
```
