Task 4

```python
arr1 = np.array([1,2,3,4])
arr2 = np.array([2,4,6,8])
sum = np.add(arr1,arr2)
product = np.multiply(arr1,arr2)
sine = np.sin(arr2)
print(sum,product,sine)
```

Task 6

```python
arr = np.arange(25).reshape(5,5)
print(arr)
arr[np.arange(5),np.arange(5)] = 0
rows, cols = np.indices(arr.shape)
print(np.indices(arr.shape))
arr[rows != cols] *= 2
print(arr)
```
